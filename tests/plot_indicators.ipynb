```python
from btopt.plots import CandlestickChart  # noqa
from btopt.data.dataloader import CSVDataLoader, YFDataloader  # noqa
from btopt import Engine
from btopt.data.timeframe import Timeframe

import pandas as pd
from pathlib import Path
import pandas_ta as ta
```

```python
engine = Engine()

start_date = "2024-04-01"
end_date = "2024-04-30"

symbol = "BTCUSDT"
timeframe = "1m"
dataloader = CSVDataLoader(symbol, "1m", start_date=start_date, end_date=end_date)
# engine.resample_data(dataloader, timeframe)
engine.add_data(dataloader)
df = engine._dataview.data[symbol][Timeframe(timeframe)]
```

```python
data_path = Path("/Users/jerryinyang/Code/database/data")
df = pd.read_parquet(data_path / "daily/EURUSD.parquet")
df.set_index(pd.to_datetime(df["time"]), drop=True, inplace=True)
df = df[
    (df.index >= pd.to_datetime("2024-04-01"))
    & (df.index < pd.to_datetime("2024-04-30"))
]
df
```

```python
vwap = ta.vwap(df["high"], df["low"], df["close"], df["volume"])
```

```python
chart = CandlestickChart()
chart.add_ohlcv_data(df)
chart.add_indicator(
    "VWAP",
    vwap,
    plot_type="line",
    style="solid",
)
```
